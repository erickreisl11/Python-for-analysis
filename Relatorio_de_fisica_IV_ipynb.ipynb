```
import numpy as np
import matplotlib.pyplot as plt


x=np.array([0.085,0.096,0.107,0.118,0.130,0.141])
y=np.array([60,68,76,84,92,100])

fig, ax = plt.subplots(1, 1 , figsize=(10,8))

ax.set_ylabel('Corrente (A)' ,color='blue' , size=12 )
ax.set_xlabel('Intensidade (%)' ,color='blue' , size=12 )

ax.set_title('Intensidade vs Corrente' , color='blue', size=15)

n=1
c = np.polyfit(x, y, n)
p = np.poly1d(c)
xp = np.linspace(x[0], x[x.size-1], 100)

d=np.absolute(y-p(x))
tamanho=d.size
l=[]
t=0

ax.plot(x, p(x), '-',x,y,'o')

ax.annotate('y=712.332x - 0.379',
            xy=(1, 0), xycoords='axes fraction',
            xytext=(-50, 200), textcoords='offset pixels',size=16,
            horizontalalignment='right',
            verticalalignment='bottom')
```
